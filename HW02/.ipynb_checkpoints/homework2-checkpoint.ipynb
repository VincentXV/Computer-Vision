## Homework 2
1. Input image from gear-inspection-system-1200x600.png.
2. The objective is to detect missing teeth using a sequence of morphological filters. The size of gears are fixed, but the position of the gears and missing teeth are random. There could be more than one missing teeth on each gear.
3. Design a set of structure elements.
4. Apply morphological filters using the structure elements.
5. Mark the defected gears using <b>red rectangles</b> on the output images.
6. Write a simple report in a separate cell.
7. Upload your Jupyter code file (*.ipynb)

Input

output

```python
#gear tooth inspection using morphological filter

import cv2
import numpy as np

output = cv2.imread('gear-inspection-system-1200x600.png')
img = cv2.imread('gear-inspection-system-1200x600.png', cv2.IMREAD_GRAYSCALE)
retval, img_a = cv2.threshold(img, 90, 255, cv2.THRESH_BINARY)

kernel_dot = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
kernel_dot1 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(20,20))
kernel_dot2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(7,7))
kernel_dot3 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(2,2))
kernel_hole = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(130,130))
kernel_hole_e = cv2.erode(kernel_hole, kernel_dot, iterations=1)
kernel_hole_ring = kernel_hole - kernel_hole_e
kernel_gear = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(50,50))

gear_body = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(15,15))
gear_tooth = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,9))

img_b = cv2.erode(img_a, kernel_hole_ring, iterations=1)
img_c = cv2.dilate(img_b, kernel_hole_e, iterations=1)
img_d = cv2.bitwise_or(img_a, img_c)

img_e = cv2.morphologyEx(img_d, cv2.MORPH_CLOSE, kernel_dot2)
img_e1 = cv2.erode(img_e, kernel_hole_ring, iterations=1)
img_e2 = cv2.dilate(img_e1, kernel_gear, iterations=3)
img_e3 = cv2.erode(img_e2, gear_body, iterations=1)
img_e = img_e2 - img_e3


img_f = cv2.bitwise_and(img_a, img_e)
img_g = cv2.dilate(img_f, kernel_dot, iterations=1)
img_g = cv2.morphologyEx(img_g, cv2.MORPH_OPEN, kernel_dot)
retval, img_g = cv2.threshold(img_g, 10, 255, cv2.THRESH_BINARY)

img_h0 = cv2.dilate(img_g, kernel_dot2, iterations=5)

img_h1 = abs(img_e - img_h0)
img_h2 = cv2.erode(img_h1, gear_tooth, iterations=1)
img_h3 = cv2.erode(img_h2, kernel_dot3, iterations=1)
img_h4 = cv2.dilate(img_h3, kernel_hole, iterations=1)
img_h4 = cv2.morphologyEx(img_h4, cv2.MORPH_OPEN, kernel_hole_e)
#img_h4 = cv2.morphologyEx(img_h4, cv2.MORPH_OPEN, kernel_dot1)
img_h = img_h4.copy()
retval, img_h = cv2.threshold(img_h, 250, 255, cv2.THRESH_BINARY)

num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(img_h)

print("labels:"+str(num_labels))

for i in range(1, num_labels):
    x = stats[i, cv2.CC_STAT_LEFT]
    y = stats[i, cv2.CC_STAT_TOP]
    w = stats[i, cv2.CC_STAT_WIDTH]
    h = stats[i, cv2.CC_STAT_HEIGHT]
    area = stats[i, cv2.CC_STAT_AREA]
    center_x, center_y = centroids[i]
    if area > 100:
        # 畫出外接矩形 (綠色)
        cv2.rectangle(output, (x, y), (x + w, y + h), (0, 255, 0), 2)
        print((x, y), (x + w, y + h))
        # 畫出形心 (紅色小圓點)
        cv2.circle(output, (int(center_x), int(center_y)), 4, (0, 0, 255), -1)
        # 標註標籤編號與面積
        cv2.putText(output, f"ID:{i}, Area:{area}", (x, y - 10), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 1)

#cv2.imshow('input', img)
#cv2.imshow('(a)', img_a)
#cv2.imshow('(b)', img_b)
#cv2.imshow('(c)', img_c)
#cv2.imshow('(d)', img_d)
#cv2.imshow('(e1)', img_e1)
#cv2.imshow('(e2)', img_e2)
#cv2.imshow('(e3)', img_e3)
#cv2.imshow('(e)', img_e)
#cv2.imshow('(f)', img_f)
#cv2.imshow('(g)', img_g)
#cv2.imshow('(h0)', img_h0)
#cv2.imshow('(h1)', img_h1)
#cv2.imshow('(h2)', img_h2)
#cv2.imshow('(h3)', img_h3)
#cv2.imshow('(h4)', img_h4)
cv2.imshow('(h)', img_h)
cv2.imshow('final', output)
cv2.waitKey()
cv2.destroyAllWindows()
```
