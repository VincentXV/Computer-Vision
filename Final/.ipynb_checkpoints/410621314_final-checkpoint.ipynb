```python
#solve wiiplay find mii game_1 : "find_this_mii"

import cv2
import numpy as np

scaling_factor = 0.5
cap = cv2.VideoCapture("WiiPlay.mp4")
frame_seq = 4820
cap.set(cv2.CAP_PROP_POS_FRAMES , frame_seq)

ret, frame = cap.read()
cur_frame = cv2.resize(frame, None, fx=scaling_factor, fy=scaling_factor)
height, width = cur_frame.shape[0:2]
show_img = np.copy(cur_frame)
mouse_pressed = False
y0 = x0 = w = h = 0

#function
def mouse_callback(event, _x, _y, flags, param):
    global show_img, x0, y0, w, h, mouse_pressed
    if event == cv2.EVENT_LBUTTONDOWN:
        mouse_pressed = True
        x0, y0 = _x, _y
        show_img = np.copy(cur_frame)
    elif event == cv2.EVENT_MOUSEMOVE:
        if mouse_pressed:
            show_img = np.copy(cur_frame)
            cv2.rectangle(show_img, (x0, y0), (_x, _y), (0, 255, 0), 2)
    elif event == cv2.EVENT_LBUTTONUP:
        mouse_pressed = False
        w, h = abs(_x - x0), abs(_y - y0)

cv2.namedWindow('image')
cv2.setMouseCallback('image', mouse_callback)

while True:
    cv2.imshow('image', show_img)
    k = cv2.waitKey(1)
    if k == ord('a') and not mouse_pressed:
        if w*h > 0:
            break
cv2.destroyAllWindows()

template = np.copy(cur_frame[y0:y0+h, x0:x0+w])
print(template.shape)
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR','cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
show_img = np.copy(cur_frame)
index = 4

while True:
    frame_seq += 1
    if frame_seq > 5000:
        frame_seq = 4820
    cap.set(cv2.CAP_PROP_POS_FRAMES , frame_seq)
    status_cap, frame = cap.read()
    if not status_cap:
        break
    cur_frame = cv2.resize(frame, None, fx=scaling_factor, fy=scaling_factor)
    cv2.imshow('find_this_mii',show_img)
    
    if 0 <= index < len(methods):
        method = methods[index]
        res = cv2.matchTemplate(cur_frame, template, eval(method))
        res = cv2.normalize(res, None, 0, 1, cv2.NORM_MINMAX)
        res = cv2.normalize(res, None, 0, 1, cv2.NORM_MINMAX)
        res_img = np.copy(res)
        
        min_val, max_val, min_loc, max_loc, = cv2.minMaxLoc(res)
        if(index >= methods.index('cv2.TM_SQDIFF')):
            cv2.rectangle(res_img, (min_loc[0] - int(w/2), min_loc[1] - int(h/2)), (min_loc[0] + int(w/2), min_loc[1] + int(h/2)), (0, 0, 255), 3)
            cv2.rectangle(cur_frame, (min_loc[0] - int(w/2), min_loc[1] - int(h/2)), (min_loc[0] + int(w/2), min_loc[1] + int(h/2)), (0, 0, 255), 3)
        else:
            cv2.rectangle(res_img, (min_loc[0] - int(w/2), min_loc[1] - int(h/2)), (min_loc[0] + int(w/2), min_loc[1] + int(h/2)), (0, 0, 255), 3)
            cv2.rectangle(cur_frame, (min_loc[0] - int(w/2), min_loc[1] - int(h/2)), (min_loc[0] + int(w/2), min_loc[1] + int(h/2)), (0, 0, 255), 3)
            
        res_img = cv2.resize(res_img, (width, height))*255
        res_img = cv2.cvtColor(res_img, cv2.COLOR_GRAY2BGR).astype(np.uint8)
        cv2.putText(res_img, method, (0, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
        show_img = np.copy(cur_frame)
        show_img = np.hstack((show_img, res_img))
        
    k = cv2.waitKey(1)
    if k == 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
#solve wiiplay find mii game_2 : "find_two_look_alike"

import cv2
import numpy as np

scaling_factor = 0.5
cap = cv2.VideoCapture("WiiPlay.mp4")
frame_seq =2200 #2180

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')


while True:
    frame_seq += 1
    if frame_seq > 2380:
        frame_seq =2200 #2180
    cap.set(cv2.CAP_PROP_POS_FRAMES , frame_seq)
    status_cap, frame0 = cap.read()
    if not status_cap:
        break
    frame = cv2.resize(frame0, None, fx=scaling_factor, fy=scaling_factor)

    #detect pedestrian
    hog = cv2.HOGDescriptor()
    hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
    #locations, weights = hog.detectMultiScale(frame, padding = (8,8), scale = 1.1, finalThreshold = 1.5)
    locations, weights = hog.detectMultiScale(frame, hitThreshold = 0, winStride=(1,1), groupThreshold = 1.0)
    #detect face
    face_rects = face_cascade.detectMultiScale(frame, scaleFactor=1.03, minNeighbors=8)
    
    #draw rects
    for (x,y,w,h) in face_rects:
        cv2.rectangle(frame, (x,y), (x+w,y+h), (255,0,0), 3)
    for loc in locations:
        cv2.rectangle(frame, (loc[0], loc[1]),(loc[0]+loc[2], loc[1]+loc[3]), (0, 255, 0), 3)
    
    
    cv2.imshow("find_two_look_alike", frame)
    k = cv2.waitKey(1)
    if k == 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
#solve wiiplay find mii game_3 : "find_the_fastest_character"

import cv2
import numpy as np

scaling_factor = 0.5
cap = cv2.VideoCapture("WiiPlay.mp4")
frame_seq = 2480 #2505
cap.set(cv2.CAP_PROP_POS_FRAMES , frame_seq)

res, frame0= cap.read()
frame = cv2.resize(frame0, None, fx=scaling_factor, fy=scaling_factor)


face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face_rects = face_cascade.detectMultiScale(frame, scaleFactor=1.04, minNeighbors=5)

lens = len(face_rects)
bbox = [0]*lens
tracker = [0]*lens
status_tracker = np.zeros(lens)
status_tracker = status_tracker.astype('bool')

i = 0
for (x,y,w,h) in face_rects:
    cv2.rectangle(frame, (x,y), (x+w,y+h), (0,255,0), 3)
    bbox[i] = (x, y, w, h)
    tracker[i] = cv2.TrackerMIL_create()
    #print(type(tracker[i].init(frame, bbox)))
    #print(type(status_tracker[i]))
    status_tracker[i] = tracker[i].init(frame, bbox[i])
    i = i + 1
cv2.imshow("first frame", frame)
cv2.waitKey()
cv2.destroyAllWindows()


while True:
    frame_seq += 1
    if frame_seq > 2600:
        frame_seq = 2480
    cap.set(cv2.CAP_PROP_POS_FRAMES , frame_seq)
    status_cap, frame0 = cap.read()
    if not status_cap:
        break
    frame = cv2.resize(frame0, None, fx=scaling_factor, fy=scaling_factor)
    i = 0
    run = np.arange(lens)
    for i in run:
        if status_tracker[i]:
            status_tracker[i], bbox[i] = tracker[i].update(frame)
        if status_tracker[i]:
            x, y, w, h = [int(j) for j in bbox[i]]
            cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 3)
        '''
        else:
            cv2.putText(frame, "Tracking failure detected", (0, 80), cv2.FONT_HERSHEY_SIMPLEX,2,(0,0,255),2)
        '''
    
    cv2.imshow("find_the_fastest_character", frame)
    k = cv2.waitKey(1)
    if k == 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
#solve wiiplay find mii game_4 : "find_two_odds"

import cv2
import numpy as np

#average = 0
def display_flow(img, flow, stride=10):
    #count = 0
    for index in np.ndindex(flow[::stride, ::stride].shape[:2]):
        pt1 = tuple(i*stride for i in index)
        delta = flow[pt1].astype(np.int32)[::-1]
        pt2 = tuple(pt1 + 2*delta)
        if 2 <= cv2.norm(delta) <= 10:
            #count = count + 1
            cv2.arrowedLine(img, pt1[::-1], pt2[::-1], (255,0,0), 1, cv2.LINE_AA, 0, 0.1)
        #print(count)
    norm_opt_flow = np.linalg.norm(flow, axis=2)
    norm_opt_flow = cv2.normalize(norm_opt_flow, None, 0, 1, cv2.NORM_MINMAX)
    cv2.imshow('optical flow', img)
    cv2.imshow('optical flow magnitude', norm_opt_flow)
    k = cv2.waitKey(1)
    if k == 27:
        return 1
    else:
        return 0

scaling_factor = 0.5
cap = cv2.VideoCapture("WiiPlay.mp4")
frame_seq =1700 #1650
cap.set(cv2.CAP_PROP_POS_FRAMES , frame_seq)
_ , prev_frame = cap.read()
prev_frame = cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY)
prev_frame = cv2.resize(prev_frame, None, fx=scaling_factor, fy=scaling_factor)
first_frame = True

while True:
    frame_seq += 4
    if frame_seq > 1800:
        frame_seq =1700 #1650
    cap.set(cv2.CAP_PROP_POS_FRAMES , frame_seq)
    status_cap, frame0 = cap.read()
    if not status_cap:
        break
    frame = cv2.resize(frame0, None, fx=scaling_factor, fy=scaling_factor)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    if first_frame:
        opt_flow = cv2.calcOpticalFlowFarneback(prev_frame, gray, None, 0.5, 5, 15, 10, 5, 1.1, cv2.OPTFLOW_FARNEBACK_GAUSSIAN)
        first_frame = False
    else:
        opt_flow = cv2.calcOpticalFlowFarneback(prev_frame, gray, opt_flow, 0.5, 5, 15, 10, 5, 1.1, cv2.OPTFLOW_USE_INITIAL_FLOW)
    prev_frame = np.copy(gray)
    if display_flow(frame, opt_flow):
        break;
    
    #cv2.imshow("find_two_odds", frame)
    k = cv2.waitKey(1)
    if k == 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```
