```python
#inpainting

import numpy as np
import cv2

img = cv2.imread('chess_board_wood2.png')
print(img.shape)
rows, cols = img.shape[:2]
mask = np.zeros((rows, cols, 1), dtype=np.uint8)
x = (int) (cols / 2)
y = (int) (rows / 2)
cv2.rectangle(mask, (x-160, y-22), (x-30, y+25), 255, cv2.FILLED);
cv2.rectangle(mask, (x+30, y-22), (x+90, y+25), 255, cv2.FILLED);
cv2.rectangle(mask, (x+100, y-22), (x+160, y+25), 255, cv2.FILLED);
dst_ns = cv2.inpaint(img, mask, 1, cv2.INPAINT_NS)  
dst_te = cv2.inpaint(img, mask, 1, cv2.INPAINT_TELEA)
cv2.imshow('input',img)
cv2.imshow('mask',mask)
cv2.imshow('dst1',dst_ns)
cv2.imshow('dst2',dst_te)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#face detector
 
import cv2
import numpy as np
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
cap = cv2.VideoCapture('wiiplay.mp4')
scaling_factor = 0.5
while True:
    ret, frame = cap.read()
    frame = cv2.resize(frame, None, fx=scaling_factor, fy=scaling_factor, interpolation=cv2.INTER_AREA)
    face_rects = face_cascade.detectMultiScale(frame, scaleFactor=1.3, minNeighbors=3)
    for (x,y,w,h) in face_rects:
        cv2.rectangle(frame, (x,y), (x+w,y+h), (0,255,0), 3)
    cv2.imshow('Face Detector', frame)
    c = cv2.waitKey(1)
    if c == 27:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
#image vignetting using separable filters (2D = 1D * 1D)

import cv2
import numpy as np

img = cv2.imread('lena.png')
rows, cols = img.shape[:2]
# generating vignette mask using Gaussian kernels
kernel_x = cv2.getGaussianKernel(cols,100)
kernel_y = cv2.getGaussianKernel(rows,100)
#2D = 1D * 1D
kernel = kernel_y * kernel_x.T
#mask = 255 * kernel / np.linalg.norm(kernel)
mask = cv2.normalize(kernel, None, 0, 1, cv2.NORM_MINMAX)

output = np.copy(img)
# applying the mask to each channel in the input image
for i in range(3):
    output[:,:,i] = output[:,:,i] * mask
    
cv2.imshow('Original', img)
cv2.imshow('Vignette', output)
cv2.imwrite('Vignette.png', output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
